```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('/kaggle/input/rsna-2023-abdominal-trauma-detection/train.csv')
train
```

```python
features = [col for col in train.columns if col != 'patient_id']
features
```

```python
submission = pd.read_csv('/kaggle/input/rsna-2023-abdominal-trauma-detection/sample_submission.csv')
submission[features] = train[features].mean().tolist()

for feature in features:
    if feature.split('_')[1] == 'low' or feature == 'bowel_injury':
        submission[feature] *= 4
    elif feature.split('_')[1] == 'high':
        submission[feature] *= 6
    elif feature.split('_')[1] == 'injury' and feature != 'bowel_injury':
        submission[feature] *= 28
        

submission.to_csv('submission.csv', index=False)
```
